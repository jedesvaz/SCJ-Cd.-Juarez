## Unzipping files

```python
#Linear regression
from zipfile import ZipFile

def unzip():
    zipfile="C:\\Users\\HP Pavilion\\Downloads\\archive.zip"
    
    with ZipFile(file=zipfile, mode="r", allowZip64=True) as file:
        archivo=file.open(name=file.namelist()[0],mode="r")
        print(archivo.read())
        archivo.close()
        
        browsing= "C:\\Users\\HP Pavilion\\Downloads"
        print("Uzipping files")
        file.extractall(path=browsing)
        print("Files unzipped")
unzip()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("C:\\Users\\HP Pavilion\\Downloads\\winequality-red.csv", delimiter=",")# skip_header=0)
print(data)
#data.ndim
#data = np.array()
```

```python
data=data.rename({'citric acid': 'citric_acid','fixed acidity':'fixed_acidity','residual sugar':'residual_sugar','total sulfur dioxide':'total_sulfur_dioxide'}, axis=1)
data
```

```python
quality=np.array(data.quality)
quality
alcohol=np.array(data.alcohol)
ph=np.array(data.pH)
citric_acid=np.array(data.citric_acid)

#np.transpose(data, axes=None)
```

```python
x=citric_acid
y=ph

#create basic scatterplot
plt.plot(x, y, 'o')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x, y, 1)

#add linear regression line to scatterplot 
plt.plot(x, m*x+b)
plt.xlabel('cido ctrico (g/dm3)')
plt.ylabel('pH')
plt.title('Modelo de regresin lineal (cido ctrico-pH)')
#plt.savefig('accitrico_ph.png')
#print('Slope: ',m)
#print('Intercept: ',b)

print(f'Equacin:Y= {m:.1f}x + {b:.1f} ')
```

```python
x=alcohol
y=ph

#create basic scatterplot
plt.plot(x, y, 'o')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x, y, 1)

#add linear regression line to scatterplot 
plt.plot(x, m*x+b, color='black')
plt.xlabel('Alcohol (% vol.)')
plt.ylabel('pH')
plt.title('Modelo de regresin lineal (Alcohol-pH)')
#plt.savefig('alcopH.png')
print(f'Equacin:Y= {m:.2f}x + {b:.1f} ')
```
